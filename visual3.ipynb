```python
import pandas as pd
import matplotlib.pyplot as plt

# 
data = {
    'Year': [2023, 2022],
    '10th': [-0.4, 5],
    '20th': [0.9, 3.2],
    '30th': [2.9, 1.9],
    '40th': [1.8, 2.4],
    '50th (Median)': [1.7, 2.6],
    '60th': [1.4, 2.3],
    '70th': [0.9, 2.8],
    '80th': [0.7, 2.2],
    '90th': [-0.3, 0]
}

df = pd.DataFrame(data)

# 
plt.figure(figsize=(10, 6))
percentiles = df.columns[1:]  # 
for percentile in percentiles:
    plt.plot(df['Year'], df[percentile], label=percentile)

plt.title('Real Change in Monthly Household Income from Work Per Household Member')
plt.xlabel('Year')
plt.ylabel('Income Change')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 
plt.figure(figsize=(10, 6))
bottom = None
for percentile in percentiles:
    if bottom is None:
        plt.bar(df['Year'], df[percentile], label=percentile)
        bottom = df[percentile]
    else:
        plt.bar(df['Year'], df[percentile], bottom=bottom, label=percentile)
        bottom += df[percentile]

plt.title('Real Change in Monthly Household Income from Work Per Household Member')
plt.xlabel('Year')
plt.ylabel('Income Change')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 
data_2022 = {
    'Percentile': ['10th', '20th', '30th', '40th', '50th (Median)', '60th', '70th', '80th', '90th'],
    'Income Change': [-0.4, 0.9, 2.9, 1.8, 1.7, 1.4, 0.9, 0.7, -0.3]
}

data_2023 = {
    'Percentile': ['10th', '20th', '30th', '40th', '50th (Median)', '60th', '70th', '80th', '90th'],
    'Income Change': [5, 3.2, 1.9, 2.4, 2.6, 2.3, 2.8, 2.2, 0]
}

# 2022
plt.figure(figsize=(10, 6))
plt.barh(data_2022['Percentile'], data_2022['Income Change'], color='skyblue')
plt.title('Real Change in Monthly Household Income from Work Per Household Member (2022)')
plt.xlabel('Income Change')
plt.ylabel('Percentile')
plt.xlim(-1, 5)
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()

# 2023
plt.figure(figsize=(10, 6))
plt.barh(data_2023['Percentile'], data_2023['Income Change'], color='lightgreen')
plt.title('Real Change in Monthly Household Income from Work Per Household Member (2023)')
plt.xlabel('Income Change')
plt.ylabel('Percentile')
plt.xlim(-1, 5)
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()
```
