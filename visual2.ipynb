```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 创建DataFrame
data = {
    'Year': [2023, 2022],
    'Resident Households Average': [12178, 11480],
    'Resident Households Median': [9329, 8615],
    'Resident Employed Households Average': [13958, 13124],
    'Resident Employed Households Median': [10869, 10099]
}

df = pd.DataFrame(data)

# 打印DataFrame，确保数据正确
print(df)

# 绘制直方图
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Resident Households Average', bins=10, kde=True, color='skyblue', label='Resident Households Average')
sns.histplot(data=df, x='Resident Employed Households Average', bins=10, kde=True, color='green', label='Resident Employed Households Average')
plt.title('Distribution of Household Income (Average)')
plt.xlabel('Monthly Household Income (SGD)')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# 绘制箱线图
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Resident Households Average', 'Resident Employed Households Average']], palette='pastel')
plt.title('Box Plot of Household Income (Average)')
plt.xlabel('Category')
plt.ylabel('Monthly Household Income (SGD)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
file_path = "/Users/mac/Downloads/17862_副本.csv"
data = pd.read_csv(file_path)

# 打印数据集的前几行，以确保正确加载
print(data.head())

# 创建条形图
plt.figure(figsize=(10, 6))
bar_width = 0.25

# 计算每种类型的住房所占比例
total_hdb = data['Total HDB2/']
hdb_dwellings = data['HDB Dwellings']
condo_apartments = data['Condominiums & Other Apartments']
landed_properties = data['Landed Properties']

# 计算堆叠条形图的位置
years = data['Year']
index = range(len(years))

# 绘制条形图
plt.bar(index, hdb_dwellings, bar_width, label='HDB Dwellings')
plt.bar(index, condo_apartments, bar_width, bottom=hdb_dwellings, label='Condominiums & Other Apartments')
plt.bar(index, landed_properties, bar_width, bottom=hdb_dwellings + condo_apartments, label='Landed Properties')

# 添加图例和标签
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Resident Households By Type of Dwelling')
plt.xticks(index, years)
plt.legend()

# 显示图表
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 数据
data = {
    'Year': [2023, 2022],
    'Total_HDB2': [77.8, 77.9],
    'HDB_1_2_Room_Flats': [7, 6.7],
    'HDB_3_Room_Flats': [17, 17.2],
    'HDB_4_Room_Flats': [31.3, 31.4],
    'HDB_5_Room_Executive_Flats': [22.4, 22.6],
    'Total_HDB': [100, 100],  # 这里我们添加了一个总数，因为它不在数据中
    'Condominiums_Other_Apartments': [17.2, 17],
    'Landed_Properties': [4.8, 4.9]
}

# 创建DataFrame
df = pd.DataFrame(data)

# 创建堆叠条形图
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = df['Year']

# 绘制条形图
plt.bar(index, df['HDB_1_2_Room_Flats'], bar_width, label='HDB 1- & 2- Room Flats')
plt.bar(index, df['HDB_3_Room_Flats'], bar_width, bottom=df['HDB_1_2_Room_Flats'], label='HDB 3-Room Flats')
plt.bar(index, df['HDB_4_Room_Flats'], bar_width, bottom=df['HDB_1_2_Room_Flats'] + df['HDB_3_Room_Flats'], label='HDB 4-Room Flats')
plt.bar(index, df['HDB_5_Room_Executive_Flats'], bar_width, bottom=df['HDB_1_2_Room_Flats'] + df['HDB_3_Room_Flats'] + df['HDB_4_Room_Flats'], label='HDB 5-Room & Executive Flats')
plt.bar(index, df['Condominiums_Other_Apartments'], bar_width, bottom=df['Total_HDB'], label='Condominiums & Other Apartments')
plt.bar(index, df['Landed_Properties'], bar_width, bottom=df['Total_HDB'] + df['Condominiums_Other_Apartments'], label='Landed Properties')

# 添加图例和标签
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Resident Households By Type of Dwelling')
plt.xticks(index, df['Year'])
plt.legend()

# 显示图表
plt.tight_layout()
plt.show()
```

```python
# 饼图
plt.figure(figsize=(8, 8))

# 绘制饼图
sizes = df.iloc[0, 2:]  # 仅使用2023年的数据
labels = df.columns[2:]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# 添加标题
plt.title('Resident Households Distribution in 2023')

# 显示图表
plt.axis('equal')
plt.show()
```
