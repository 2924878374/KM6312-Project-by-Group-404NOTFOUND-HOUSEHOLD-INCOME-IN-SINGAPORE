```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
file_path = "/Users/mac/Downloads/17868/17868-table_副本.csv"
data = pd.read_csv(file_path)

# 提取月收入范围和对应的百分比
income_ranges = data.iloc[1:, 0].values
percentages_2022 = data.iloc[1:, 2].values
percentages_2023 = data.iloc[1:, 1].values

# 设置图形大小
plt.figure(figsize=(10, 6))

# 绘制堆积条形图
plt.barh(income_ranges, percentages_2022, color='skyblue', label='2022', alpha=0.6)
plt.barh(income_ranges, percentages_2023, color='salmon', left=percentages_2022, label='2023', alpha=0.6)

# 添加标签和标题
plt.xlabel('Percentage')
plt.ylabel('Income Range')
plt.title('Monthly Household Income Distribution (2022 and 2023)')

# 显示图例
plt.legend()

# 设置y轴刻度标签
plt.yticks(range(len(income_ranges)), income_ranges)

# 显示图形
plt.show()
```

```python
 # 设置图形大小
plt.figure(figsize=(10, 8))

# 绘制堆积条形图
plt.bar(income_ranges, percentages_2022, color='skyblue', label='2022', alpha=0.6)
plt.bar(income_ranges, percentages_2023, color='salmon', bottom=percentages_2022, label='2023', alpha=0.6)

# 添加标签和标题
plt.xlabel('Income Range')
plt.ylabel('Percentage')
plt.title('Monthly Household Income Distribution (2022 and 2023)')

# 显示图例
plt.legend()

# 旋转x轴刻度标签
plt.xticks(rotation=90)

# 显示图形
plt.show()
```

```python
# 设置图形大小
plt.figure(figsize=(12, 8))

# 绘制2022年数据的条形图
plt.bar(income_ranges, percentages_2022, color='skyblue', label='2022', alpha=0.6)

# 绘制2023年数据的条形图
plt.bar(income_ranges, percentages_2023, color='salmon', bottom=percentages_2022, label='2023', alpha=0.6)

# 添加标签和标题
plt.xlabel('Income Range')
plt.ylabel('Percentage')
plt.title('Monthly Household Income Distribution by Year')

# 显示图例
plt.legend()

# 旋转x轴刻度标签
plt.xticks(rotation=90)

# 显示图形
plt.show()

```

```python
# 设置图形大小
plt.figure(figsize=(12, 8))

# 绘制2022年数据的折线图
plt.plot(income_ranges, percentages_2022, marker='o', linestyle='-', color='skyblue', label='2022')

# 绘制2023年数据的折线图
plt.plot(income_ranges, percentages_2023, marker='o', linestyle='-', color='salmon', label='2023')

# 添加标签和标题
plt.xlabel('Income Range')
plt.ylabel('Percentage')
plt.title('Monthly Household Income Distribution by Year')

# 显示图例
plt.legend()

# 旋转x轴刻度标签
plt.xticks(rotation=90)

# 显示网格线
plt.grid(True)

# 显示图形
plt.show()
```
